```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("/content/Sample - Superstore.xlsx")
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df['Postal Code'].fillna(0, inplace=True)
```

```python
sns.histplot(df['Sales'])
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
df['tota cost']=df['Sales']-df['Profit']
```

```python
df['profit_margin']=df['Profit']/df['Sales']
```

```python
df['sales per unit']=df['Sales']/df['Quantity']
```

```python
sns.barplot(x='sales per unit',y='profit_margin',data=df)
plt.show()
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
